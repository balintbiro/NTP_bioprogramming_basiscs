Two types of differentiation will be briefly discussed in the followings:
- symbolic --> when you have a formula that you have to differentiate
- numeric --> when you have data; you can plot a graph and you have to differentaite that data (it is more common than the previous one)

```python
#importations
import scipy as sp
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

Sympy is the only tool that we have not used before. It is a great library for mathematical operations. You can do basically every calculus stuff with it.

```python
#one of the most powerful features of sympy that it makes it possible to define mathematical variables
x=smp.symbols('x',real=True)#it is a good practice to define the variables as real numbers
x
```

```python
#after defining the variable it is possible to execute different mathematical operations on them
x**4
```

```python
#applying a well known differentiation rule in python
f=x**9
dfdx=smp.diff(f,x)
dfdx
```

```python
#you can differentiate very cheeky formulas
x,a,b,c=smp.symbols('x a b c')
f=smp.exp(-a*smp.sin(x**2))*smp.sin(b**x)*smp.log(c*smp.sin(x)**2/x)
f
```

```python
#derivative of the above formula. This is a long creepy derivative that you dont wanna do by hand...
dfdx=smp.diff(f,x)
dfdx
```

As we have discussed in the "introduction" ppt, the derivative of a function is gonna be a function itself but it is also possible to have a derivative of a function at some point. In this case you should replace variables with the numbers provided.

```python
dfdx.subs([(x,4),(a,1),(b,2),(c,3)])
```

```python
#after solving the above expression we will this number as the derivative of our function with the above
#variables replaced by the above numbers
dfdx.subs([(x,4),(a,1),(b,2),(c,3)]).evalf()
```

In real world scenarios, it is more common to have measured datapoints. We can plot a figure of them and then take the derivative of it.

```python
#visualize a simple vector
x=np.arange(1,21)
y=np.random.normal(size=20)
plt.plot(x,y,'o--')
```

```python
#get the derivative of it. Remember, because we have differentiated the whole function (without specifying
#any special points), the result of the differentiation is gonna be another function
dydx=np.gradient(y,x)
plt.plot(x,y,'o--')
plt.plot(x,dydx,'o--')
```
