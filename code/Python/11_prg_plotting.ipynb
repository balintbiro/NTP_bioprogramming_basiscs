```python
#common importation, from that point you can use the plt
import matplotlib.pyplot as plt
```

```python
#this is a very common way to use the plotter in python
#fig is the "external frame"
#axs are the subplots in a matrix
#by tweaking the ncols, nrows pair you can adjust the number of subplots
fig, axs=plt.subplots(ncols=2, nrows=2)
```

```python
#since the axs are the subplots in a matrix, you can refer to individual subplots as matrix indexing
fig, axs=plt.subplots(ncols=2, nrows=2)
axs[0,0].plot([1,2,3,4],[1,2,3,4])#simple line plot
```

```python
#it is also possible to have variables and just use their names
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=2, nrows=2)
axs[0,0].plot(x,y)
```

```python
#the subplots function can be used in cases when you need just one plot
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=1, nrows=1)
axs.plot(x,y)
```

```python
#barplot
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=2, nrows=2)
axs[0,0].plot(x,y)
axs[0,1].bar(x,y)
```

```python
#scatterplot
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=2, nrows=2)
axs[0,0].plot(x,y)
axs[0,1].bar(x,y)
axs[1,0].scatter(x,y)
```

```python
#pieplot
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=2, nrows=2)
axs[0,0].plot(x,y)
axs[0,1].bar(x,y)
axs[1,0].scatter(x,y)
axs[1,1].pie(x)
```

```python
#modify different aspects of a subplot
x=[1,2,3,4]
y=[1,2,3,4]
fig, axs=plt.subplots(ncols=2, nrows=2)

axs[0,0].plot(x,y)
axs[0,0].grid()
axs[0,0].set_ylabel('Magassag')

axs[0,1].bar(x,y)
axs[1,0].scatter(x,y)
axs[1,1].pie(x)
```
