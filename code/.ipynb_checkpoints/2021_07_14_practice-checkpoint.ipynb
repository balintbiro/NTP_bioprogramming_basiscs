```python
sequence = ''
with open('../data/6ltj.pdb') as raw_file:
    content = raw_file.readlines()
    for line in content:
        if line[0] == 'A'and line[1] == 'T' : 
            if line.rsplit()[0] == 'ATOM' and line.rsplit()[4] == 'D':
                residue = line.rsplit()[3]
                sequence += residue
```
